# Finetune SMI-TED for property serve

> Run IBM's full finetune (encoder + AE encoder + `Net`) on a **GPU**, then drop the checkpoint where MAX export expects it.

This notebook is **not** part of the `mat_gram01` nbdev export (`00`–`07`). It wraps the vendored IBM finetune code so you can train ESOL / BBBP / lipo and hand the `.pt` to `export_finetune_to_max.py`.

```text
GPU finetune (.pt) → max_ports/finetune_ckpts/{task}/ → pixi run export-finetune → serve-*
```


## Where this lives

| Path | Role |
|------|------|
| This notebook (`nbs/08_finetune.ipynb`) | Interactive finetune driver |
| `vendor/.../finetune/` | IBM training code + MoleculeNet CSVs |
| `model_assets/ibm-research_materials.smi-ted/smi-ted-Light_40.pt` | Pretrained weights |
| `finetune_ckpts/{esol,bbbp,lipo}/` | **Save the best `.pt` here** |
| `scripts/export_finetune_to_max.py` | `.pt` → MAX property asset |

**Needs a GPU.** CPU will run but 500 epochs is impractical.


## 1. Config — pick a task


````python
from __future__ import annotations

import os
import shutil
import sys
from pathlib import Path

# --- edit these ---
TASK = "esol"  # "esol" | "bbbp" | "lipo"
MAX_EPOCHS = 500  # IBM default; use 2–5 for a smoke test
N_BATCH = 32
LR_START = 3e-5
START_SEED = 0
# ------------------

REPO = Path.cwd().resolve()
if REPO.name == "nbs":
    REPO = REPO.parent
assert (REPO / "materials_smi_ted").is_dir(), f"run from max_ports/ or max_ports/nbs/, not {REPO}"

VENDOR_FINETUNE = (
    REPO.parent / "vendor" / "ibm_materials" / "models" / "smi_ted" / "finetune"
)
LIGHT = VENDOR_FINETUNE / "smi_ted_light"
ASSETS = REPO / "model_assets" / "ibm-research_materials.smi-ted"
CKPT_OUT = REPO / "finetune_ckpts" / TASK

TASKS = {
    "esol": {
        "script": "finetune_regression",
        "data_root": VENDOR_FINETUNE / "moleculenet" / "esol",
        "dataset_name": "esol",
        "measure_name": "measured log solubility in mols per litre",
        "loss_fn": "rmse",
        "target_metric": "rmse",
        "n_output": 1,
        "train_decoder": 1,
        "ibm_ckpt_dir": LIGHT / "esol" / "checkpoints_esol",
    },
    "bbbp": {
        "script": "finetune_classification",
        "data_root": VENDOR_FINETUNE / "moleculenet" / "bbbp",
        "dataset_name": "bbbp",
        "measure_name": "p_np",
        "loss_fn": "crossentropy",
        "target_metric": "roc-auc",
        "n_output": 2,
        "train_decoder": 0,
        "ibm_ckpt_dir": LIGHT / "bbbp" / "checkpoints_bbbp",
    },
    "lipo": {
        "script": "finetune_regression",
        "data_root": VENDOR_FINETUNE / "moleculenet" / "lipophilicity",
        "dataset_name": "lipophilicity",
        "measure_name": "y",
        "loss_fn": "rmse",
        "target_metric": "rmse",
        "n_output": 1,
        "train_decoder": 1,
        "ibm_ckpt_dir": LIGHT / "lipo" / "checkpoints_lipophilicity",
    },
}

assert TASK in TASKS, f"TASK must be one of {sorted(TASKS)}"
cfg = TASKS[TASK]
print("REPO         ", REPO)
print("TASK         ", TASK)
print("MAX_EPOCHS   ", MAX_EPOCHS)
print("data         ", cfg["data_root"])
print("save (IBM)   ", cfg["ibm_ckpt_dir"])
print("copy to      ", CKPT_OUT)
````

## 2. Prerequisites

Link pretrained weights + `fast_transformers` into the IBM finetune tree (same layout as their README).


````python
import torch

print("cuda available:", torch.cuda.is_available())
if torch.cuda.is_available():
    print("device:", torch.cuda.get_device_name(0))
else:
    print("WARNING: no GPU — finetune will be very slow")


def link_or_copy(src: Path, dest: Path) -> None:
    dest.parent.mkdir(parents=True, exist_ok=True)
    if dest.exists() or dest.is_symlink():
        if dest.resolve() == src.resolve():
            print("ok", dest)
            return
        dest.unlink()
    try:
        dest.symlink_to(src.resolve())
        print("linked", dest, "->", src)
    except OSError:
        if src.is_dir():
            shutil.copytree(src, dest)
        else:
            shutil.copy2(src, dest)
        print("copied", dest)


pt = ASSETS / "smi-ted-Light_40.pt"
vocab = ASSETS / "bert_vocab_curated.txt"
if not pt.is_file():
    raise FileNotFoundError(f"missing {pt}; run: pixi run setup-model")
if not vocab.is_file():
    raise FileNotFoundError(f"missing {vocab}; run: pixi run setup-model")

link_or_copy(pt, LIGHT / "smi-ted-Light_40.pt")
link_or_copy(vocab, LIGHT / "bert_vocab_curated.txt")

ft_src = (
    REPO.parent
    / "vendor"
    / "ibm_materials"
    / "models"
    / "smi_ted"
    / "smi_ted_light"
    / "fast_transformers"
)
ft_dst = LIGHT / "fast_transformers"
if not ft_dst.exists():
    if not ft_src.is_dir():
        raise FileNotFoundError(f"missing {ft_src}")
    link_or_copy(ft_src, ft_dst)
else:
    print("ok", ft_dst)

assert (VENDOR_FINETUNE / "finetune_regression.py").is_file()
assert cfg["data_root"].is_dir(), cfg["data_root"]
CKPT_OUT.mkdir(parents=True, exist_ok=True)
print("prerequisites ok")
````

## 3. Run IBM finetune

Uses the same hyperparameters as `run_finetune_{esol,bbbp,lipo}.sh`. Checkpoints are written under the IBM task folder; we copy the best file into `finetune_ckpts/` afterward.


````python
from argparse import Namespace

# IBM imports resolve relative to finetune/
os.chdir(VENDOR_FINETUNE)
if str(VENDOR_FINETUNE) not in sys.path:
    sys.path.insert(0, str(VENDOR_FINETUNE))

ibm_ckpt_dir = cfg["ibm_ckpt_dir"]
ibm_ckpt_dir.mkdir(parents=True, exist_ok=True)

config = Namespace(
    n_batch=N_BATCH,
    n_layer=12,
    n_head=12,
    n_embd=768,
    max_len=202,
    d_dropout=0.1,
    dropout=0.1,
    lr_start=LR_START,
    lr_multiplier=1,
    max_epochs=MAX_EPOCHS,
    num_feats=32,
    smi_ted_version="v1",
    model_path=str(LIGHT),
    ckpt_filename="smi-ted-Light_40.pt",
    data_root=str(cfg["data_root"]),
    dataset_name=cfg["dataset_name"],
    measure_name=cfg["measure_name"],
    checkpoints_folder=str(ibm_ckpt_dir),
    loss_fn=cfg["loss_fn"],
    target_metric=cfg["target_metric"],
    n_output=cfg["n_output"],
    save_ckpt=1,
    save_every_epoch=0,
    start_seed=START_SEED,
    train_decoder=cfg["train_decoder"],
    restart_filename=None,
)

print(config)
````

````python
# This cell trains — can take hours at MAX_EPOCHS=500 on one GPU.
if cfg["script"] == "finetune_regression":
    from finetune_regression import main as finetune_main
else:
    from finetune_classification import main as finetune_main

finetune_main(config)
````

## 4. Copy best checkpoint → `finetune_ckpts/`

IBM names files like `smi-ted-Light-Finetune_seed0_esol_epoch=..._valloss=....pt`. We copy the newest (best-so-far; trainer deletes older when `save_every_epoch=0`).


````python
pts = sorted(ibm_ckpt_dir.glob("*Finetune*.pt"), key=lambda p: p.stat().st_mtime)
if not pts:
    raise FileNotFoundError(f"no *Finetune*.pt under {ibm_ckpt_dir}")

best = pts[-1]
dest = CKPT_OUT / best.name
shutil.copy2(best, dest)
print("best checkpoint:", best)
print("copied to:      ", dest)
print("size_mb:        ", round(dest.stat().st_size / 1e6, 1))
````

## 5. Export to MAX (optional, same machine)

After the `.pt` is in `finetune_ckpts/`, build the property asset:

```bash
cd max_ports
pixi run export-finetune -- --checkpoint finetune_ckpts/esol/YOUR.pt --task esol
pixi run serve-esol
pixi run predict-finetuned -- --task esol --smiles CCO
```

Or run the export from this notebook:


````python
os.chdir(REPO)
sys.path.insert(0, str(REPO / "scripts"))

from export_finetune_to_max import export_finetune

out = export_finetune(
    checkpoint=dest,
    task=TASK,  # type: ignore[arg-type]
    output_dir=REPO / "model_assets" / f"smi-ted-{TASK}",
    base_assets=ASSETS,
)
print("MAX asset dir:", out)
````

## Notes

- **ESOL / lipo:** regression, `n_output=1`, `train_decoder=1` (matches IBM shells).
- **BBBP:** classification, `n_output=2`, `train_decoder=0`. MAX property serve currently assumes `n_output=1` logit for BBBP — if you need the IBM 2-class head in MAX, extend the export/graph next.
- Smoke test: set `MAX_EPOCHS = 2` before the train cell.
- Kernel: use the `max_ports` pixi env (`pixi shell` then `jupyter notebook`), or any env with IBM finetune deps (PyTorch + CUDA, rdkit, etc.).
